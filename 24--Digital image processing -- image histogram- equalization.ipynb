```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from skimage import data
```

```python
#gray scale image histogram with the help of matplotlib
img=cv.imread(r"D:\output image\tiger.jpg",0)
cv.waitKey()
cv.destroyAllWindows()
plt.hist(img.ravel(),256,[0,256])
plt.title("histogram")
plt.xlabel('pixel value')
plt.show()
```

```python
#color scale image histogram
image=img=cv.imread(r"D:\output image\tiger.jpg")
cv.waitKey()
cv.destroyAllWindows()


#histogram from each color
color=('b','g','r')
for i,col in enumerate(color):
    hist=cv.calcHist([img],[i],None,[256],[0,256])
plt.plot(hist,color=col)
plt.title('Color histogram')
plt.xlim([0,256])
plt.show()
```

```python
#Histogram equalizer
#gray scale image histogram
img=cv.imread(r"D:\output image\tiger.jpg",0)
hist=cv.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)


#histogram after equalizing the image
image_hist=cv.equalizeHist(img)
hist=cv.calcHist([image_hist],[0],None,[256],[0,256])
plt.plot(hist)

cv.imshow("Real image",img)
cv.imshow("After equalizing ",image_hist)
cv.waitKey()
cv.destroyAllWindows()
```

```python
#color image histogram....

image=cv.imread(r"D:\output image\Skin Cancer Image 1.png")
b,g,r=cv.split(image)
cv.imshow("color image", image)
cv.waitKey()
cv.destroyAllWindows()
hist=cv.calcHist([b],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([g],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([r],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#color image histogram  eqalizer with hsv color spacing.

image=cv.imread(r"D:\output image\Skin Cancer Image 1.png")
b,g,r=cv.split(image)
image_hsv=cv.cvtColor(image, cv.COLOR_BGR2HSV)
h,s,v=cv.split(image_hsv)
v=cv.equalizeHist(v)
merge_hsv=cv.merge((h,s,v))
bgr_enhance=cv.cvtColor(merge_hsv,cv.COLOR_HLS2BGR)
cv.imshow("color image", image)
cv.imshow("After equalize", bgr_enhance)
cv.waitKey()
cv.destroyAllWindows()
b,g,r=cv.split(bgr_enhance)
hist=cv.calcHist([b],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([g],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([r],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
#color image histogram  eqalizer with rgb color spacing.
image=cv.imread(r"D:\output image\Skin Cancer Image 1.png")
b,g,r=cv.split(image)

b=cv.equalizeHist(b)
g=cv.equalizeHist(g)
r=cv.equalizeHist(r)
merge_bgr=cv.merge((b,g,r))
cv.imshow("color image", image)
cv.imshow("After equalize", merge_bgr)
cv.waitKey()
cv.destroyAllWindows()
hist=cv.calcHist([b],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([g],[0],None,[256],[0,256])
plt.plot(hist)
hist=cv.calcHist([r],[0],None,[256],[0,256])
plt.plot(hist)
plt.title(' RGB Color histogram eqalization')
plt.xlabel("Channel RGB" )
plt.ylabel("Color intensity frequency" )
plt.show()
```
