```python
from skimage import data
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

image=cv.imread(r"D:\output image\tiger.jpg",0)
image=cv.medianBlur(image,5)      #image smooth
 

 #Global thresholding
res1,thresh=cv.threshold(image,127,255, cv.THRESH_BINARY)    

#adaptive thresholding
th2=cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,5,5)       #GAUSSIAN
th3=cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,5,5)           #Mean


title=['Original Image', 'Global thresholding', 'Adaptive mean thresholding', 'Adaptive gaussion thresholding']
images=[image,thresh,th2,th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(title[i])
    plt.xticks([]),plt.yticks([])
plt.show()

```

```python
cv.imshow("Image",image)
cv.imshow("GAUSSIAN",th2)
cv.imshow("Mean",th3)
cv.waitKey()
cv.destroyAllWindows()
```
