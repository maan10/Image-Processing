Within the scikit-image package, there are several sample images provided in the data module. Letâ€™s say we want to load a single image to perform a few experiments.  Instead of using an external image, we can simply load one of the images provided within the package!


```python
from skimage.io import imread, imshow
from skimage import data
image=data.astronaut()          #data contain multiple image like coin, cat, astronaut
img1=data.camera()
imshow(img1)   
imshow(image)
```

What if you want to load an image from your machine instead of the ones provided in the package?For this, we can use the imread
function from skimage.

```python
# converting image into gray scale.......
import matplotlib.pyplot 
gray_img=imread('E:\\practining image\\flower.png',as_gray=True)
print(gray_img.shape)
imshow(gray_img)
```

```python
#reading colored image

import matplotlib.pyplot 
gray_img=imread('E:\\practining image\\flower.png',as_gray=False)   # parameter value difference
print(gray_img.shape)
imshow(gray_img)
```
