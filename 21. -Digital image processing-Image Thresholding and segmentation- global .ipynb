```python
from skimage import data
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
orignal_image=cv.imread(r"D:\output image\cat_rgb.jpg",0)     #gray
orignal_image_1=cv.imread(r"D:\output image\Tiger.jpg",0)   


#tresholding implement:

#image 1
res1, thres1= cv.threshold(orignal_image,127,255,cv.THRESH_BINARY)              #Type 1
res1, thres2= cv.threshold(orignal_image,127,255,cv.THRESH_BINARY_INV)          #Type 2
res1, thres3= cv.threshold(orignal_image,127,255,cv.THRESH_TRUNC)               #Type 3
res1, thres4= cv.threshold(orignal_image,127,255,cv.THRESH_TOZERO)              #Type 4
res1, thres5= cv.threshold(orignal_image,127,255,cv.THRESH_TOZERO_INV)          #Type 5



#image 2
res1, thres6= cv.threshold(orignal_image_1,127,255,cv.THRESH_BINARY)              #Type 1
res1, thres7= cv.threshold(orignal_image_1,127,255,cv.THRESH_BINARY_INV)          #Type 2
res1, thres8= cv.threshold(orignal_image_1,127,255,cv.THRESH_TRUNC)               #Type 3
res1, thres9= cv.threshold(orignal_image_1,127,255,cv.THRESH_TOZERO)              #Type 4
res1, thres10= cv.threshold(orignal_image_1,127,255,cv.THRESH_TOZERO_INV)          #Type 5



Title=['Original image', 'Original image 1', ' Binary', 'Binary','Binary inver', 'Binary inver','trunc', 'trunc',
       'Tozero','Tozero', 'Tozero inver',  'Tozero inver']

image=[orignal_image, orignal_image_1, thres1,thres6, thres2,thres7, thres3, thres8, thres4,thres9, thres5,thres10]

for i in range(12):
    plt.subplot(3,4, i+1), plt.imshow(image[i], 'gray')
    plt.title(Title[i])
    plt.xticks([]), plt.yticks([])

plt.show()
cv.waitKey()
cv.destroyAllWindows()


```
