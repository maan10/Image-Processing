```python
from skimage import data
import cv2 as cv

image=data.cat()
gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
hsv=cv.cvtColor(image,cv.COLOR_BGR2HSV)
cv.imshow("Camera",image)
cv.imshow("gray",gray)
cv.imshow("hsv",hsv)
cv.imwrite(r"D:\output image\cat_rgb.jpg",image)
cv.imwrite(r"D:\output image\cat_gray.jpg",gray)
cv.imwrite(r"D:\output image\cat_hsv.jpg",hsv)
#image property
print("----------------------------------------------------------------------------------------------------------------------")
print('\timage RGB')
print("Image Shape: {} ".format(image.shape))
print("Image Height in pixel: {} ".format(image.shape[0]))
print("Image width in pixel: {} ".format(image.shape[1]))
print("Image channel {}".format(image.shape[2]))
print("Image size in total pixel : {}".format(image.size))   #height* width *channel
print("TYpe of image", type(image))

#pexel read:

#in RGM ---->BGR
pexel=image[200,221]
print(pexel)
(B,G,R)=pexel
print(f"Image Color Space in RGB: R {R}, G {G}, R{R} ")
#--------------------------------------------------------------------------------------------------------------------------------


print("----------------------------------------------------------------------------------------------------------------------")
print('\tGray Scale image ')
print("Image Shape: {} ".format(gray.shape))
print("Image Height in pixel: {} ".format(gray.shape[0]))
print("Image width in pixel: {} ".format(gray.shape[1]))
#print("Image channel {}".format(image.shape[2]))
print("Image size in total pixel : {}".format(gray.size))   #height* width *channel
print("TYpe of image", type(gray))
#pexel read:

#in RGM ---->BGR
pexel=gray[200,221]
print(pexel)

#--------------------------------------------------------------------------------------------------------------------------------


print("----------------------------------------------------------------------------------------------------------------------")
print('\timage hsv')
print("Image Shape: {} ".format(hsv.shape))
print("Image Height in pixel: {} ".format(hsv.shape[0]))
print("Image width in pixel: {} ".format(hsv.shape[1]))
print("Image channel {}".format(hsv.shape[2]))
print("Image size in total pixel : {}".format(hsv.size))   #height* width *channel
print("TYpe of image", type(hsv))
#pexel read:

#in RGM ---->BGR

pexel=hsv[200,221]
print(pexel)
(H,S,V)=pexel
print(f"Image Color Space in RGB: H {H}, S {S}, V{V} ")
#--------------------------------------------------------------------------------------------------------------------------------










cv.waitKey(0) 
cv.destroyAllWindows()

```
